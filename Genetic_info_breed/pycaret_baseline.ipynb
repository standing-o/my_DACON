```python
from pycaret.classification import *

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import sklearn
import sys
import seaborn as sns
import random as rn
import os


import warnings
%matplotlib inline
warnings.filterwarnings(action='ignore')
```

```python
# reproducibility
def set_seed(seed_num):
  # tf.random.set_seed(seed_num)
  np.random.seed(seed_num)
  rn.seed(seed_num)
  os.environ['PYTHONHASHSEED']=str(seed_num)

seed_num = 42
set_seed(seed_num)
```

```python
train_df = pd.read_csv('dataset/train.csv')
test_df = pd.read_csv('dataset/test.csv')
info = pd.read_csv('dataset/snp_info.csv')
```

```python
drop_cols = ['id', 'father', 'mother', 'gender']

train_y = train_df['class']

train_df = train_df.drop(drop_cols, axis = 1)
test_df = test_df.drop(drop_cols, axis = 1)
```

```python
train_x = train_df.copy()
test_x = test_df.copy()

print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
```

```python
setup_clf = setup(data=train_x, target='class', train_size=0.8,
                  session_id=seed_num)
```

```python
models()
```

```python
model = compare_models(sort='F1',fold=10, n_select=10)
```

```python
tuned_model = [tune_model(i) for i in model]
```

```python
tuned_model
```

```python
blended = blend_models(estimator_list = tuned_model,
                       fold = 10,
                       method = 'hard',
                       optimize='F1',
                       )
```

```python
final_model = finalize_model(blended)
evaluate_model(final_model)
```

```python
from pycaret.utils import check_metric

prediction = predict_model(final_model, data=test_x)
```
